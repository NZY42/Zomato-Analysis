REMOVING NEW , - , \ from rate column

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
newDf['rate'] = newDf['rate'].apply(handlerate)
newDf['rate'].head()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',' , '')
        return float(value)
    else:
        return float(value)
    
newDf['cost2plates'] = newDf['cost2plates'].apply(handlecomma)
newDf['cost2plates'].unique()
                              
```

```python
newDf.head()
```

Cleaning Rest Type Column 

```python
newDf['rest_type'].value_counts()
```

```python
rest_types = newDf['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_typeslessthan1000 = rest_types[rest_types<1000]
rest_typeslessthan1000
```

```python
def handle_rest_type(value):
    if value in rest_typeslessthan1000:
        return 'others'
    else:
        return value 
newDf['rest_type']=newDf['rest_type'].apply(handle_rest_type)
newDf['rest_type'].value_counts()    
```

```python
newDf.head()
```

Cleaning Location Column


```python
newDf['location'].value_counts()
```

```python
location = newDf['location'].value_counts(ascending=False)
location_lessthan300 = location[location<300]

def handle_location(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return value
    
newDf['location'] = newDf['location'].apply(handle_location)
newDf['location'].value_counts()
    
```

```python
newDf.head()
```

cleaning cuisines column


```python
cuisines = newDf['cuisines'].value_counts(ascending=False)
cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
newDf['cuisines']=newDf['cuisines'].apply(handle_cuisines)
newDf['cuisines'].value_counts()
```
